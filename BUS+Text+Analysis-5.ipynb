```python
#Chirag Narayanakere

from urllib.request import urlopen
import json
import pandas
from pandas.io.json import json_normalize

#this is the url that we've already determined is safe and legal to scrape from
request = 'http://api.crossref.org/journals/00222429/works?filter=from-pub-date:2000-01-01&select=DOI&rows=1000'
request2 = 'http://api.crossref.org/journals/00222429/works?filter=from-pub-date:2000-01-01&select=DOI&rows=1000&offset=1000'

#here, we fetch the content from the url, using the requests library
response = urlopen(request, timeout=5)
response2 = urlopen(request2, timeout=5)

#here, we store the response (as a json object)
dois = response.read()
dois2 = response2.read()

#here, we read the json object into a dictionary
test = pandas.read_json(dois)
test2 = pandas.read_json(dois2)

#here, we convert the dictionary to a dataframe to allow for easier manipulation
master = pd.DataFrame.from_dict(test.iloc[1, 0])
master2 = pd.DataFrame.from_dict(test2.iloc[1, 0])

#combined dataframe
master_final = pd.concat([master, master2]).reset_index(drop=True)

#adding the base url to all the doi values
for i in range(0, len(master_final)):
    master_final['DOI'][i] = "https://journals.sagepub.com/doi/pdf/" + master_final['DOI'][i]

#final dataframe
master_final
```
