```python
#Chirag Narayanakere

#df1 = first 1000 entries dataframe

#df = last 42 entries dataframe

#df_final = combined dataframe

#Here, we're just importing both Beautiful Soup and the Requests library
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import operator

#first 1000 results
page_link1 = 'http://api.crossref.org/journals/00222429/works?filter=from-pub-date:2000-01-01&select=DOI&rows=1000'
#last 42 results
page_link2 = 'http://api.crossref.org/journals/00222429/works?filter=from-pub-date:2000-01-01&select=DOI&rows=1000&offset=1000'

# this is the url that we've already determined is safe and legal to scrape from.
page_response1 = requests.get(page_link1, timeout=5)
page_response2 = requests.get(page_link2, timeout=5)

# here, we fetch the content from the url, using the requests library
page_content1 = BeautifulSoup(page_response1.content, "html.parser")
page_content2 = BeautifulSoup(page_response2.content, "html.parser")

# DF1 ----------------------------------------------------------------------------------------
#separate by commas
holder = page_content1.getText().split(",")

#create dataframe to manipulate
df1 = pd.DataFrame({'col': holder}).drop([0, 1, 2, 3, 4, 1005, 1006, 1007])
df1 = df1.reset_index(drop=True)

#edit column value for row 0
df1['col'][0] = df1['col'][0][10:].replace(":", "").replace('"', "").lower()
df1['col'][0] = df1['col'][0][:3] + '/' + df1['col'][0][3:]
df1['col'][0] = "https://journals.sagepub.com/" + df1['col'][0][0:4] + "pdf/" + df1['col'][0][8:15] + df1['col'][0][16:-3]

#editing column values for all remaining rows
for i in range(1, 1000):
    df1['col'][i] = df1['col'][i].replace(":", "").replace('"', "").lower()
    df1['col'][i] = df1['col'][i][:4] + '/' + df1['col'][i][4:]
    df1['col'][i] = "https://journals.sagepub.com/" + df1['col'][i][1:5] + "pdf" + df1['col'][i][4:12] + df1['col'][i][13:-1]

df1['col'][999] = df1['col'][999][:-1]
# end DF1 ------------------------------------------------------------------------------------

# DF -----------------------------------------------------------------------------------------
#separate by commas
holder2 = page_content2.getText().split(",")

#create dataframe to manipulate
df = pd.DataFrame({'col': holder2}).drop([0, 1, 2, 3, 4])
df = df.reset_index(drop=True)

#edit column value for row 0
df['col'][0] = df['col'][0][10:].replace(":", "").replace('"', "").lower()
df['col'][0] = df['col'][0][:3] + '/' + df['col'][0][3:10] + df['col'][0][11:-1]
df['col'][0] = "https://journals.sagepub.com/" + df['col'][0][0:4] + "pdf/" + df['col'][0][4:16] + df['col'][0][16:-3]

#editing column values for all remaining rows
for i in range(1, 28):
    df['col'][i] = df['col'][i].replace(":", "").replace('"', "").lower()
    df['col'][i] = df['col'][i][:4] + '/' + df['col'][i][4:]
    df['col'][i] = "https://journals.sagepub.com/" + df['col'][i][1:5] + "pdf" + df['col'][i][4:12] + df['col'][i][13:-1]

#edit column value for row 28
df['col'][28] = df['col'][28][8:].replace(":", "").replace('"', "").lower()
df['col'][28] = "https://journals.sagepub.com/doi/pdf/" + df['col'][28][:7] + df['col'][28][8:-2]

#drop unnecesary rows
df = df.drop([29, 30, 31])
# end DF -------------------------------------------------------------------------------------

#display full dataframe output
pd.set_option('display.max_colwidth', -1)

#combine dataframes into one, reset index
df_final = pd.concat([df1, df])
df_final = df_final.reset_index(drop=True)

df_final['col'][0] = df_final['col'][0][:37] + "10.1509/jmp.67.2.123.18610"

#display dataframe
df_final

```
